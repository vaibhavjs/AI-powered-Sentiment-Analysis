<font size="3"><b>2.1 Data Preprocessing</b></font>
<br><br>
Sentiment analysis is a method that uses natural language processing and machine learning to analyze text and understand the emotions expressed in it. To make this analysis more effective, we need to preprocess and clean up the dataset. This process helps us identify the important words and phrases that carry the information we're looking for. By doing this, we can gain valuable insights from the text and understand the sentiment behind it more easily.

Let's start diving into the actual code now - 

<br>
<font size="3"><b>2.1.1 Importing necessary libraries</b></font>
<br>

```python
# Importing all the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from wordcloud import WordCloud
from bs4 import BeautifulSoup
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Ignoring warnings
import warnings 
warnings.filterwarnings('ignore')
```

<br>
<font size="3"><b>2.1.2 Loading & Reading the dataset</b></font>
<br>

```python
# Loading the dataset as a dataframe

df=pd.read_csv("IMDB_Dataset.csv")
df
```

<br>
<font size="3"><b>2.1.3 Dataframe shape, checking missing values & sentiment counts </b></font>
<br>

```python
df.shape
```

```python
df.isnull().sum()
```

No null values present in the dataframe.

```python
df['sentiment'].value_counts()
```

Equal distribution of occurences of sentiment labels. Balanced dataset.

```python
df.head()
```

<br>
<font size="3"><b>2.1.4 Actual Preprocessing begins: Removing HTML tags</b></font>
<br>

```python
# Removing HTML tags
def removehtml(text):
    soup=BeautifulSoup(text)
    return soup.get_text()

df['review']=df['review'].apply(removehtml)
```

<br>
<font size="3"><b>2.1.5 Converting into lowercase </b></font>
<br>

```python
# Converting the dataframe into lowercase

df['review']=df['review'].str.lower()
```

<br>
<font size="3"><b>2.1.6 Replacing everything except alphabets & numbers with space </b></font>
<br>

```python
df['review']=df['review'].str.replace(r'[^a-zA-Z0-9]',' ')
```

```python
df.head()
```

Now, we're starting to see a little cleaned-up data.

<br>
<font size="3"><b>2.1.7 Removing stopwords & implementing stemming </b></font>
<br>

```python
# Creating a set sw containing stopwords in English
sw=set(stopwords.words('english'))


# Initializing the SnowballStemmer with English
snow=SnowballStemmer('english')


# Stemming implementation
def stemming(text):
    text=' '.join([snow.stem(word) for word in text.split() if word not in sw])
    return text

df['review']=df['review'].apply(stemming)
```

```python
df.head()
```

<br>
<font size="3"><b>2.1.8 Label coding 0 & 1 </b></font>
<br>

```python
df['sentiment'].replace({'negative':0,'positive':1},inplace=True)
```

```python
df.head()
```

By converting the textual labels into numerical codes, we make it easier for machine learning models to process and analyze the sentiment data. Numerical labels enable our models to perform various computations and comparisons more efficiently and effectively.

<br>
<font size="3"><b>Word cloud for positive sentiment reviews </b></font>
<br>

```python
# Visualization of most frequently occuring words in the positive reviews

positive=df['review'][df['sentiment']==1]
spamcloud=WordCloud(width=1200,height=800,background_color='black',max_words=250).generate(' '.join(positive))

plt.figure(figsize=(12,8))
plt.imshow(spamcloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

<br>
<font size="3"><b>Word cloud for negative sentiment reviews </b></font>
<br>

```python
# Visualization of most frequently occuring words in the negative reviews

negative=df['review'][df['sentiment']==0]
spamcloud=WordCloud(width=1200,height=800,background_color='black',max_words=250).generate(' '.join(negative))

plt.figure(figsize=(12,8))
plt.imshow(spamcloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

Such word cloud visualisations provides us with a quick overview of the common themes and sentiments expressed in positive / negative movie reviews.

<br>
<font size="3"><b>2.1.9 Text Vectorization using TF-IDF & Extracting sentiment labels </b></font>
<br>

```python
# Text Vectorization with TF-IDF
review=df['review']
tfidf=TfidfVectorizer(ngram_range=(1,3))
review=tfidf.fit_transform(review)


# Extracting sentiment labels
sentiment=df['sentiment']
```

TF-IDF assigns importance scores to words based on their frequency in a specific review and across all reviews in the dataset. The TfidfVectorizer is used to convert the textual data into a numerical format, enabling machine learning models to process and analyze the reviews effectively.

<br>
<font size="3"><b>2.1.10 Splitting the dataset into training and testing sets </b></font>
<br>

```python
xtrain,xtest,ytrain,ytest=train_test_split(review,sentiment,test_size=0.3,random_state=11)
```

Allocating 30% of the preprocessed data for testing, while the remaining 70% will be used for training.

<br><br><br>
<font size="3"><b>2.2 Baseline Performance </b></font>
<br>

In this section, we will explore and evaluate the performance of the Multinomial Naive Bayes model, which we consider as our baseline approach for sentiment analysis. 

To evaluate the performance of the Multinomial Naive Bayes model, we employ several key metrics. Firstly, the accuracy score is calculated by comparing the predicted sentiment labels with the actual sentiment labels in the test dataset. The accuracy score indicates the percentage of correctly classified samples, offering a measure of overall performance.

Next, we present the confusion matrix, which provides a comprehensive view of the model's predictions. The confusion matrix displays the true positive, true negative, false positive, and false negative predictions, allowing us to assess the model's effectiveness in differentiating positive and negative sentiments.

Furthermore, the classification report furnishes us with detailed insights into the model's precision, recall, and F1-score for both positive and negative sentiments. Precision represents the proportion of correctly predicted positive (or negative) samples among all predicted positive (or negative) samples. Recall, on the other hand, indicates the proportion of correctly predicted positive (or negative) samples among all actual positive (or negative) samples. The F1-score is the harmonic mean of precision and recall, providing a balanced measure of the model's performance.

By thoroughly evaluating the Multinomial Naive Bayes model using these metrics, we gain valuable insights into its strengths and limitations. This serves as the foundation for comparing the performance of more complex models, such as the Logistic Regression approach, which we will explore in the subsequent section. Through this comparative analysis, we aim to identify the most effective model for sentiment analysis and contribute to advancing the field's understanding of natural language processing and machine learning techniques.

```python
# Create and train the MultinomialNaiveBayesClassifier

multi_naive_bayes_model = MultinomialNB()
multi_naive_bayes_model.fit(xtrain, ytrain)
```

```python
# Predict sentiments

preds1=multi_naive_bayes_model.predict(xtest)
```

```python
# Performance Evaluaton 

print('Accuracy score :')
print(accuracy_score(ytest, preds1))
print('==================================================')
print('Confusion Matrix :')
print(confusion_matrix(ytest, preds1))
print('==================================================')
print('Classification Report :')
print(classification_report(ytest, preds1))
print('==================================================')
```

We will analyze these performance metrics in-depth later in the evaluation section. For now, let's focus on generating the similar results for our main model - Logistic Rgression and then we will compare it with this baseline MNB model.

<br><br><br>
<font size="3"><b>2.3 Classification approach </b></font>
<br>

Our classification approach involves utilizing the Logistic Regression model as an alternative to the baseline Multinomial Naive Bayes model.

The reason we chose Logistic Regression as an alternative to the baseline Multinomial Naive Bayes model is because of its simplicity and interpretability. Logistic Regression works well in cases where the relationship between features and the target variable is roughly linear. It also provides probabilities for each sentiment class, allowing us to understand how confident the model is in its predictions.

Logistic Regression has several benefits. It performs well on smaller datasets like ours and is less prone to overfitting. Moreover, it is computationally efficient, making it suitable for real-time applications. However, it may not capture complex relationships in the data as effectively as more sophisticated models, like neural networks, and its performance may suffer if the data has significant non-linearity or high dimensionality.

In our implementation, we use the scikit-learn library for training and testing the Logistic Regression model. Our primary contribution lies in selecting and evaluating the model, interpreting the results, and drawing meaningful conclusions. By comparing the performance of Logistic Regression with the baseline Multinomial Naive Bayes model, we aim to understand which algorithm performs better in sentiment analysis tasks related to movie reviews.

```python
# Create and train the Logistic Regression model

logistic_regression_model = LogisticRegression()
logistic_regression_model.fit(xtrain, ytrain)
```

```python
# Predict sentiments

preds2=logistic_regression_model.predict(xtest)
```

```python
# Performance Evaluaton 

print('Accuracy score :')
print(accuracy_score(ytest, preds2))
print('==================================================')
print('Confusion Matrix :')
print(confusion_matrix(ytest, preds2))
print('==================================================')
print('Classification Report :')
print(classification_report(ytest, preds2))
print('==================================================')
```

The minor differences in accuracy, precision, and recall between the two models indicate that our approach can effectively match the baseline model's sentiment classification capabilities.

Through the comparison, we can confidently say that our Logistic Regression approach is promising, as it achieves a performance level very close to the baseline MNB model. These results demonstrate the effectiveness of our chosen approach in sentiment analysis tasks and establish it as a strong contender for further evaluations.
